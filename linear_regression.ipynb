### Spliting the data into training and tesing data

```python
x = df[["Age", "Gender"]]
y = df["Salary"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

#### Fiting the model

```python
reg=  LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```
